```python
from graphfla.problems import NK

nk_model = NK(n=10, k=5)
```

```python
df = nk_model.get_data()
df
```

```python
import pandas as pd
from graphfla.landscape import Landscape

X = df["config"].apply(pd.Series)
f = df["fitness"]

data_types = {x: "boolean" for x in X.columns}

landscape = Landscape(X, f, maximize=True, data_types=data_types)
```

```python
df
```

```python
ffi = landscape.ffi()
assortativity = landscape.fitness_assortativity()
autocorr = landscape.autocorrelation()
neutrality = landscape.neutrality()
ruggedness = landscape.ruggedness()
gradient_intensity = landscape.gradient_intensity()
```

```python
landscape.draw_neighborhood(node=1)
```

```python
lon = landscape.get_lon()
```

```python
landscape.draw_landscape_3d()
```

```python
landscape.all_mutation_effects()
```

```python
landscape.all_pairwise_epistasis()
```

```python
landscape.draw_epistasis()
```
